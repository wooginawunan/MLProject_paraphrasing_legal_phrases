```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
data = pd.read_csv('all_speakers.csv')
```

```python
data = data.dropna()
```

```python
print('Party1 Zero Count: ', sum(data.Democratic == 0 ))
print('Party2 Zero Count: ', sum(data.Republican == 0 ))
print('Both Zero Count: ', sum(data[data.Republican == 0].Democratic == 0 ))
```
