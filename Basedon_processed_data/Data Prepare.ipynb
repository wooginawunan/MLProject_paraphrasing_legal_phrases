```python
import pandas as pd
import numpy as np
```

```python
datapath = '/Users/apple/Desktop/Machine_Learning/project/data_audi/'
member_path = datapath+'name.tsv'
party_path = datapath+'party.tsv'
```

```python
member = pd.read_table(member_path, sep='\t', header=0)
party = pd.read_table(party_path, sep='\t', header=0)
```

```python
party = party[['MEMBER','PARTY']]
party.MEMBER = [int(x) for x in party.MEMBER]
member = member[['MEMBER','NAME']]
member = member[1:-4]
member.MEMBER = [int(x) for x in member.MEMBER]
```

```python
Member_party = pd.merge(party, member, on='MEMBER', how='right', indicator=True)
```

```python
Member_party = Member_party[Member_party['_merge'] == 'both'][['NAME','PARTY']]
```

```python
Member_party = Member_party.drop_duplicates()
```

```python
Member_party.to_csv('/Users/apple/Desktop/Machine_Learning/project/data_audi/CongressMemberParty.csv')
```

```python
def name_tranform(name):
    l = name.split(',',1)
    if len(l)==2:
        r = [l[1][1:],l[0]]
        string = ' '.join(r)
    else:
        string = name
    return string
```

```python
name_tran = []
name_NO_middle=[]

for name in Member_party.NAME:
    #print(name)
    new = name_tranform(name)
    name_tran.append(new)
    new_all = new.split(' ')
    name_NO_middle.append(' '.join([new_all[0],new_all[-1]]))
    
    
```

```python
Member_party['name_tran'] = name_tran
Member_party['name_NO_middle'] = name_NO_middle
#Member_party = Member_party.drop_duplicates()
Member_party
```
