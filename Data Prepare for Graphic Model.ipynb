```python
Speaker_A = Speaker_A.rename(columns={'F1_A1_count_vowel':'count_vowel'})
Speaker_B = Speaker_B.rename(columns={'F1_B_count_vowel':'count_vowel'})
```

```python
Speaker_A['vowel'] = list(Speaker_A.index.get_level_values('vowel'))
Speaker_B['vowel'] = list(Speaker_B.index.get_level_values('vowel'))
Speaker_A.index = Speaker_A.index.droplevel(level = 'vowel')
Speaker_B.index = Speaker_B.index.droplevel(level = 'vowel')
Speaker_A.index = Speaker_A.index.droplevel(level = 'docket_id')
Speaker_B.index = Speaker_B.index.droplevel(level = 'docket_id')
Speaker_A.index = Speaker_A.index.droplevel(level = 'aba_id')
Speaker_B.index = Speaker_B.index.droplevel(level = 'aba_id')
```

```python
Speaker_A['speaker'] = list(Speaker_A.index.get_level_values('speakerA'))
Speaker_B['speaker'] = list(Speaker_B.index.get_level_values('speakerB'))
```

```python
Speaker_A.index = range(len(Speaker_A))
Speaker_B.index = range(len(Speaker_B))
```

```python
Speaker_A

def data_transform(df,vowel_count,speaker = 'A'):
    data = {'Speaker':[],
            'count_vowel_AA':[],
            'count_vowel_AE':[],
            'count_vowel_AH':[],
            'count_vowel_AO':[],
            'count_vowel_AW':[],
            'count_vowel_AY':[],
            'count_vowel_EH':[],
            'count_vowel_ER':[],
            'count_vowel_EY':[],
            'count_vowel_IH':[],
            'count_vowel_IH':[],
            'count_vowel_IY':[],
            'count_vowel_OW':[],
            'count_vowel_OY':[],
            'count_vowel_UH':[],
            'count_vowel_UW':[]}
    row = 0
    for count in vowel_count:
        data['Speaker'].append(df.ix[row]['speaker'])
        feature_flag = {'AA':1,
                    'AE':1,
                    'AH':1,
                    'AO':1,
                    'AW':1,
                    'AY':1,
                    'EH':1,
                    'ER':1,
                    'EY':1,
                    'IH':1,
                    'IY':1,
                    'OW':1,
                    'OY':1,
                    'UH':1,
                    'UW':1}
        for i in range(count): 
            vo = df.ix[row]['vowel']
            feature_flag[vo] = 0
            #print(vo)
            data['count_vowel'+'_'+vo].append(df.ix[row]['count_vowel'])
            row = row + 1
        for vowel in feature_flag.keys():
            if feature_flag[vowel]:
                data['count_vowel'+'_'+vowel].append(0)
        
    return pd.DataFrame(data)
            
data_A = data_transform(Speaker_A,vowel_count_speakerA,speaker = 'A')
data_B = data_transform(Speaker_B,vowel_count_speakerB,speaker = 'B')
        
```
